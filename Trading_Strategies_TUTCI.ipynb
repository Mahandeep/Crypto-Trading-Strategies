```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filename = 'btcusdt_1h.csv'
data=pd.read_csv('/content/'+filename)
df1 = pd.DataFrame(data)
```

```python
def getMin(lst):
  temp = float('inf')
  for x in lst:
    temp = min(x,temp)
  return temp

def getMax(lst):
  temp = 0
  for x in lst:
    temp = max(x,temp)
  return temp
```

```python
#TURTLE TRADE CHANNEL
entryPeriod = 20
exitPeriod = 10

def popAndAppend(lst,elem):
  lst.append(elem)
  lst.pop(0)

def TUTCI(df):
  global entryPeriod,exitPeriod
  signals = []
  highPricesEntry = []
  lowPricesEntry = []
  highPricesExit = []
  lowPricesExit = []
  position = 0
  for i in range(entryPeriod):
    signals.append(0)
    highPricesEntry.append(df['high'][i])
    lowPricesEntry.append(df['low'][i])
  for i in range(exitPeriod):
    highPricesExit.append(df['high'][i])
    lowPricesExit.append(df['low'][i])
  for i in range(entryPeriod,len(df)-1):
    buySignal = df['high'][i]>=getMax(highPricesEntry)
    sellSignal = df['low'][i]<=getMin(lowPricesEntry)
    buyExit = df['low'][i]<=getMin(lowPricesExit)
    sellExit = df['high'][i]>=getMax(highPricesExit)
    popAndAppend(highPricesEntry,df['high'][i])
    popAndAppend(highPricesExit,df['high'][i])
    popAndAppend(lowPricesEntry,df['low'][i])
    popAndAppend(lowPricesExit,df['low'][i])

    if position==0:
      if buySignal:
        position=1
      elif sellSignal:
        position=-1
      signals.append(position)
    elif position==1:
      if buyExit:
        position = 0
        signals.append(-1)
      else:
        signals.append(0)
    elif position==-1:
      if sellExit:
        position=0
        signals.append(1)
      else:
        signals.append(0)
  signals.append(-1*position)
  df['signals'] = signals
```

```python
TUTCI(df1)
```

```python
df1['Index'] = range(0,len(df1))
df1 = df1[['Index','datetime','signals','open','high','low','close','volume']]
df1
```
